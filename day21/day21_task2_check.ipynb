Machine Learning Types

1️ Supervised Learning
Data has labels. The model learns from input and correct output.

2️ Unsupervised Learning
Data has no labels. The model finds hidden patterns or groups.

3️ Reinforcement Learning
The model learns by rewards and penalties while interacting with an environment.

|  Type          | Description                        | Real-World Examples                                |
| ---------------------- | ---------------------------------- | -------------------------------------------------- |
| Supervised Learning    | Data has correct answers (labels)  | Spam Email Detection, House Price Prediction       |
| Unsupervised Learning  | Finds patterns in unlabeled data   | Customer Segmentation, Market Basket Analysis      |
| Reinforcement Learning | Learns using rewards and penalties | Game Playing (Chess/Go), Self-Driving Car Training |
