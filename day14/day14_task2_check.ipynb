```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data = pd.DataFrame({
    'Age': [22, 25, 30, 35, 40, 28, 32, 45],
    'Salary': [20000, 25000, 30000, 40000, 50000, 32000, 38000, 60000]
})

print("Original Data:")
print(data)

std_scaler = StandardScaler()
data_standardized = pd.DataFrame(
    std_scaler.fit_transform(data),
    columns=data.columns
)

minmax_scaler = MinMaxScaler()
data_normalized = pd.DataFrame(
    minmax_scaler.fit_transform(data),
    columns=data.columns
)

print("\nStandardized Data:")
print(data_standardized)

print("\nNormalized Data:")
print(data_normalized)

plt.figure()
plt.hist(data['Salary'])
plt.title("Salary Before Scaling")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(data_standardized['Salary'])
plt.title("Salary After Standardization")
plt.xlabel("Scaled Salary")
plt.ylabel("Frequency")
plt.show()
```
