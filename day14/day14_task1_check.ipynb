```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame({
    'Transmission': ['Automatic', 'Manual', 'Automatic', 'Manual'],
    'Color': ['Red', 'Blue', 'Green', 'Red']
})

print("Original Data:")
print(data)

le = LabelEncoder()
data['Transmission_encoded'] = le.fit_transform(data['Transmission'])

color_encoded = pd.get_dummies(data['Color'], drop_first=True)

data_encoded = pd.concat([data.drop(['Transmission', 'Color'], axis=1), color_encoded], axis=1)

print("\nEncoded Data:")
print(data_encoded)
```
