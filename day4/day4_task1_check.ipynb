```python
# Create dictionary with three contacts
contacts = {
    "Raju": "9876543210",
    "Ramesh": "9123456780",
    "Suresh": "9988776655"
}

# Add a new contact
contacts["Satish"] = "9012345678"

# Update existing contact
contacts["Ramesh"] = "1234567890"

# Safe access using .get()
print("Lookup Raju:", contacts.get("Raju"))
print("Lookup Emma:", contacts.get("Emma", "Contact not found"))

# Iterate and print all contacts
print("\nContact List:")
for name, phone in contacts.items():
    print(f"Contact: {name} | Phone: {phone}")
```
