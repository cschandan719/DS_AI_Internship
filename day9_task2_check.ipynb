```python
import pandas as pd

grades = pd.Series([85, None, 92, 45, None, 78, 55])

print("Missing values:\n", grades.isnull())

filled_grades = grades.fillna(0)

filtered = filled_grades[filled_grades > 60]

print("\nOriginal Series:\n", grades)
print("\nFilled Series:\n", filled_grades)
print("\nScores > 60:\n", filtered)
```
