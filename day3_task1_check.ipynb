```python
# Create inventory list
inventory = ["Apples", "Bananas", "Carrots", "Dates"]

# Print current inventory
print("Current Inventory:", inventory)

# Add "Eggs" 
inventory.append("Eggs")

# Remove "Bananas" 
inventory.remove("Bananas")

# Sort inventory 
inventory.sort()

print("Final Updated Inventory:", inventory)
```
