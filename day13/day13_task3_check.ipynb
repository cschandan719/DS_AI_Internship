```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample dataset
np.random.seed(42)

df = pd.DataFrame({
    'Size': np.random.randint(800, 3000, 50),
    'Price': np.random.randint(100000, 500000, 50),
})

# Create a highly correlated feature
df['Price_Adjusted'] = df['Price'] * 1.05

# ----- Correlation Matrix -----
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix)

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ----- Find correlations > 0.8 -----
high_corr = corr_matrix[(corr_matrix > 0.8) & (corr_matrix < 1.0)]
print("\nHighly correlated variables:\n", high_corr.dropna(how='all').dropna(axis=1, how='all'))

# ----- Boxplot for outliers -----
plt.boxplot(df['Price'])
plt.title("Boxplot of Price")
plt.show()
```
