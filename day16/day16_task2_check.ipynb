```python
import pandas as pd
import numpy as np

data = {
    'Marks': [50, 55, 60, 52, 58, 61, 59, 300]  # 300 is extreme
}

df = pd.DataFrame(data)

mean = df['Marks'].mean()
std = df['Marks'].std()

print("Mean:", mean)
print("Standard Deviation:", std)

df['z_score'] = (df['Marks'] - mean) / std

print("\nData with Z-score:")
print(df)

outliers = df[abs(df['z_score']) > 3]

print("\nOutliers:")
print(outliers)
```
