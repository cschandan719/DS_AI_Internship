```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

heights = np.random.normal(170, 10, 1000)

incomes = np.random.exponential(50000, 1000)

scores = 100 - np.random.exponential(10, 1000)

def analyze(data, title):
    mean = np.mean(data)
    median = np.median(data)
    
    print(f"{title}")
    print("Mean:", round(mean,2))
    print("Median:", round(median,2))
    print("--------------------")
    
    sns.histplot(data, kde=True)
    plt.title(title)
    plt.show()

analyze(heights, "Normal Distribution (Heights)")
analyze(incomes, "Right-Skewed (Incomes)")
analyze(scores, "Left-Skewed (Easy Exam Scores)")
```
